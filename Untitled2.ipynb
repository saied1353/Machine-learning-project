```python
df['Floor_no'] = pd.to_numeric(
    df['Floor_no'].str.replace('[^0-9]', '', regex=True)
)


```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
df['Occupancy_status']=en.fit_transform(df['Occupancy_status'])
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
df['City']=en.fit_transform(df['City'])
```

```python
df['Price_in_taka']=pd.to_numeric(df['Price_in_taka'].str.replace('[^0-9]','',regex=True))
```

```python
df.info()
```

```python
import nltk
nltk.download('stopwords')
```

```python


import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
ps = PorterStemmer()
stop = stopwords.words("english")

import re

def clean_location(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9, ]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df['Location_clean'] = df['Location'].apply(clean_location)

tfidf = TfidfVectorizer(
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.8
)

location_tfidf = tfidf.fit_transform(df['Location_clean'])
location_df = pd.DataFrame(
    location_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)
df = df.drop(['Location', 'Location_clean'], axis=1)
df = pd.concat([df, location_df], axis=1)
```
